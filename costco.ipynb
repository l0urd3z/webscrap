Aquí vamos a definir la función

```python
def Buscador_Precios_Selenium_Costco(producto):
    path ="C:\webdriver3\chromedriver.exe"
    driver=webdriver.Chrome(path)
    url="https://www.costco.com.mx/search?text="+producto+""
    driver.get(url)
    
    productos= driver.find_elements_by_class_name("product-item")
    
    lista_urls=list()
    for i in range(len(productos)):
        try:
            lista_urls.append(productos[i].find_element_by_tag_name("a").get_attribute("href"))
        except: 
            lista_urls.append(np.nan)
    lista_nombres=list()
    for i in range(len(productos)): 
        try: 
            lista_nombres.append(productos[i].find_elements_by_tag_name("a")[1].text)
        except: 
            lista_nombres.append(np.nan)
            
    lista_precios=list()
    lista_promos=list()
    for i in range(len(productos)):
        try:
            lista_precios.append(productos[i].find_elements_by_class_name("original-price")[0].text)
        except: 
            lista_precios.append(np.nan)
            
    df_costco =pd.DataFrame({"nombre":lista_nombres, "url":lista_urls, "precio1":lista_precios})
    df_costco["autoservicio"]="costco"
    df_costco["marca"]=producto
    df_costco["fecha"]= time.strftime("%d/%m/%y")
    
    df_costco = df_costco[["fecha", "autoservicio", "marca", "nombre", "url", "precio1"]]
    
    #df_costco= df_costco[df_costco['nombre'].astype(str).str.contains(r'\b'.format(producto), regex=True, case=False)]
    df_costco =df_costco.reset_index(drop=True)
    
    datos_webscraper=pd.read_excel("df_costco.xlsx")
    
    datos_webscraper=pd.concat([datos_webscraper,df_costco], axis=0)
    
    datos_webscraper.to_excel("df_costco.xlsx", index=False)
    
    driver.quit()
    return df_costco
```

```python
for productos in ["juguetes%20para%20perro","camas%20para%20perro", "croquetas%20para%20perro"]:
    Buscador_Precios_Selenium_Costco(productos)
```
